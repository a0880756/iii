```python
###########把法家思想的書籍資訊都爬下來了
import requests
import pandas as pd
import json
import re
from bs4 import BeautifulSoup
import sqlalchemy
engine = sqlalchemy.create_engine("mysql+pymysql://root:fff810924@127.0.0.1:3306/?charset=utf8mb4") #創立和MYSQL的連線
engine.execute("CREATE DATABASE books CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci;")#改成utf8mb4編碼比較不會錯，建立db

headers = {
'Cookie':'customerid=192798681734WF3SWYnqw; app_ad_closed=0; __BWfp=c1510154327592x2b23da575; ai_user=6ocS7|2017-11-08T16:31:44.789Z; ASPSESSIONIDSQRRCTQQ=OPGJGLKBABDLBGNPJPPLNFMP; kingstone_=ovST/8ymQmB+fsDX26ywr2v/1TI0000; _ga=GA1.3.2067812324.1510154327; _gid=GA1.3.699490028.1510154327; _gat=1; MyCount=0; MyAmt=0; NSC_xxx_mc*80=ffffffffaf181c1d45525d5f4f58455e445a4a423660; ai_session=fTM9|1510192668569.5|1510192668569.5',
'Upgrade-Insecure-Requests':'1',
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.89 Safari/537.36'
}
##先找到最後一頁的頁碼
resp3 = requests.get('https://www.kingstone.com.tw/book/book_class.asp?class_id=HLF', headers = headers)
soup3 = BeautifulSoup(resp3.text,'html5lib')
selects = soup3.select('div.switcher > a')
# print(selects)
href = selects[-1].get('href')
# print(href)
last_page = re.findall('page=(.+)#ClassSec',href)[0]
print(last_page)

book_list =[]
for pg in range(1,int(last_page)):
    URL = 'https://www.kingstone.com.tw/book/book_class.asp?class_id=HLF&page=%s'%pg

    resp=requests.get(URL,headers = headers)
    resp.encoding = 'utf8'
    soup = BeautifulSoup(resp.text,'html5lib')

    book_names = soup.select('div.row_list > ul > li > a.anchor > span')
    authors    = soup.select('div.row_list > ul > li > span.author')
    publishers = soup.select('div.row_list > ul > li > span.publisher')
    prices     = soup.select('div.row_list > ul > li > span.price > span.sale_price')
    tags       = soup.select('div#breadCrumb > a')
    for i in range(0,20):
        book = {}
        book['bookname']  = book_names[i].text
        book['author']    = authors[i].text
        book['publisher'] = publishers[i].text
        book['price']     = prices[i].text
        book['tag']       = tags[0].text+'>'+tags[1].text+'>'+tags[2].text+'>'+tags[3].text+'>'+tags[4].text
        book_list.append(book)
df = pd.DataFrame(book_list)
df.to_sql('books_data',engine, if_exists='append', schema='books',index=False)
```
